```python
import time
import json
import threading


from azure.iot.device import IoTHubDeviceClient, Message
STRING_CONEXAO="HostName=Atividade-IOT.azure-devices.net;DeviceId=DispositivoTeste;SharedAccessKey=GEY9r94e+jOhkPJAVqLoq8vGB5j+oztsrVO8GwWZt/U="
cliente = IoTHubDeviceClient.create_from_connection_string(STRING_CONEXAO)
cliente.connect()

intervaloverde = 5
intervaloamarelo = 5
intervalovermelho = 5
modo = "a"

def tratar_msg(msg):

  global modo
  global intervaloverde
  global intervaloamarelo
  global intervalovermelho

  d1 = json.loads(msg.data)

  if d1["intervaloverde"] > 0 and d1["intervaloverde"] <= 30:
      intervaloverde = int(d1["intervaloverde"])
  else:
      print("ERRO : CONFIGURAÇÃO INVÁLIDA")

  if d1["intervaloamarelo"] > 0 and d1["intervaloamarelo"] <= 30 :
      intervaloamarelo = int(d1["intervaloamarelo"])
  else:
      print("ERRO : CONFIGURAÇÃO INVÁLIDA")

  if d1["intervalovermelho"] > 0 and d1["intervalovermelho"] <= 30:
      intervalovermelho = int(d1["intervalovermelho"])
  else:
      print("ERRO : CONFIGURAÇÃO INVÁLIDA")

  if d1["modo"] == modo:
      print("esse modo ja está em funcionamento")

  elif str(d1["modo"]) != modo:
      modo = str(d1["modo"])


cliente.on_message_received = tratar_msg




# Código Exemplo
def verde():
    print("VERDE")
    estado = json.dumps([{'Cor': 'VERDE'}])
    cliente.send_message(estado)

def amarelo():
    print("AMARELO")
    estado = json.dumps([{'Cor': 'AMARELO'}])
    cliente.send_message(estado)

def vermelho():
    print("VERMELHO")
    estado = json.dumps([{'Cor': 'VERMELHO'}])
    cliente.send_message(estado)


while True:
  if modo == "normal" :
      verde()
      time.sleep(intervaloverde)
      amarelo()
      time.sleep(intervaloamarelo)
      vermelho()
      time.sleep(intervalovermelho)

  if modo == "piscante" :
      amarelo()
      time.sleep(intervaloamarelo)
```
